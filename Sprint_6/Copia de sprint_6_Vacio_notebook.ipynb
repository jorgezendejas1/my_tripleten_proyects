# Task
Analyze the provided video game sales data ("https://practicum-content.s3.us-west-1.amazonaws.com/datasets/games.csv") to identify patterns that determine commercial success and inform the 2017 advertising campaign.

## Load and inspect the data

### Subtask:
Load the `games.csv` file into a pandas DataFrame and get a general understanding of the data, including checking for missing values and data types.


## Prepare the data

### Subtask:
Clean up column names, convert data types where necessary (e.g., `User_Score` to numeric, `Year_of_Release` to integer), handle missing values appropriately (e.g., fill with a placeholder, drop rows/columns depending on the context), and calculate the total sales for each game.


**Reasoning**:
Clean column names, convert data types, handle missing values, and calculate total sales as per the instructions.



```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')

df.dropna(subset=['year_of_release'], inplace=True)
df['year_of_release'] = df['year_of_release'].astype(int)

df.dropna(subset=['name', 'genre'], inplace=True)

df['total_sales'] = df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis=1)

display(df.head())
display(df.info())
```

## Analyze the data

### Subtask:
Look at the number of games released per year and identify the most relevant period for analysis (considering the data is up to 2016), analyze sales by platform and identify the leading platforms, analyze sales by genre and identify the most profitable genres, investigate the relationship between critic scores, user scores, and sales, and examine the distribution of sales by region (NA, EU, JP, Other).


**Reasoning**:
Calculate and display the number of games released per year, total sales per platform and genre, and total sales per region.



```python
games_per_year = df['year_of_release'].value_counts().sort_index()
print("Number of games released per year:")
display(games_per_year)

platform_sales = df.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
print("\nTotal sales per platform:")
display(platform_sales)
print("\nTop platforms:")
display(platform_sales.head())

genre_sales = df.groupby('genre')['total_sales'].sum().sort_values(ascending=False)
print("\nTotal sales per genre:")
display(genre_sales)
print("\nTop genres:")
display(genre_sales.head())

regional_sales = df[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum()
print("\nTotal sales by region:")
display(regional_sales)
```

**Reasoning**:
Create scatter plots to visualize the relationship between critic scores, user scores, and total sales, and calculate the correlation coefficients.



```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x='critic_score', y='total_sales', data=df)
plt.title('Critic Score vs. Total Sales')

plt.subplot(1, 2, 2)
sns.scatterplot(x='user_score', y='total_sales', data=df)
plt.title('User Score vs. Total Sales')

plt.tight_layout()
plt.show()

critic_corr = df['critic_score'].corr(df['total_sales'])
user_corr = df['user_score'].corr(df['total_sales'])

print(f"\nCorrelation between Critic Score and Total Sales: {critic_corr:.2f}")
print(f"Correlation between User Score and Total Sales: {user_corr:.2f}")
```

## Profile platforms

### Subtask:
Analyze sales trends for the leading platforms and determine if the average sales vary across platforms.


**Reasoning**:
Identify the top 5 platforms by total sales and filter the DataFrame to include only these platforms.



```python
top_platforms = df.groupby('platform')['total_sales'].sum().sort_values(ascending=False).head(5).index.tolist()
df_top_platforms = df[df['platform'].isin(top_platforms)].copy()
display(df_top_platforms.head())
```

**Reasoning**:
Create a pivot table to analyze yearly sales trends for each of the top platforms and plot the sales trend.



```python
platform_yearly_sales = df_top_platforms.pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')

plt.figure(figsize=(14, 8))
for platform in platform_yearly_sales.columns:
    sns.lineplot(x=platform_yearly_sales.index, y=platform_yearly_sales[platform], label=platform)

plt.title('Yearly Sales Trends for Top Platforms')
plt.xlabel('Year of Release')
plt.ylabel('Total Sales')
plt.legend(title='Platform')
plt.grid(True)
plt.show()
```

**Reasoning**:
Calculate the average total sales for each of the top platforms and conduct an ANOVA test to determine if there is a significant difference in average sales across these platforms.



```python
average_sales_per_platform = df_top_platforms.groupby('platform')['total_sales'].mean()
print("Average total sales per top platform:")
display(average_sales_per_platform)

platform_sales_list = [df_top_platforms[df_top_platforms['platform'] == platform]['total_sales'].dropna() for platform in top_platforms]
f_statistic, p_value = st.f_oneway(*platform_sales_list)

print(f"\nANOVA test results:")
print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: The average total sales vary significantly across the top platforms.")
else:
    print("Conclusion: The average total sales do not vary significantly across the top platforms.")
```

## Profile genres

### Subtask:
Analyze sales trends for different genres and determine if the average sales vary across genres.


**Reasoning**:
Calculate the total sales for each genre and sort them in descending order, then display the top genres by total sales.



```python
genre_sales = df.groupby('genre')['total_sales'].sum().sort_values(ascending=False)
print("Total sales per genre:")
display(genre_sales)
print("\nTop genres by total sales:")
display(genre_sales.head())
```

**Reasoning**:
Create a bar plot to visualize the total sales per genre.



```python
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_sales.index, y=genre_sales.values)
plt.title('Total Sales per Genre')
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Calculate the average total sales for each genre and store the total sales for each genre in separate lists or arrays for the ANOVA test. Then, perform the ANOVA test and print the results with interpretation.



```python
average_sales_per_genre = df.groupby('genre')['total_sales'].mean()
print("Average total sales per genre:")
display(average_sales_per_genre)

genre_sales_list = [df[df['genre'] == genre]['total_sales'].dropna() for genre in df['genre'].unique()]
f_statistic, p_value = st.f_oneway(*genre_sales_list)

print(f"\nANOVA test results:")
print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: The average total sales vary significantly across genres.")
else:
    print("Conclusion: The average total sales do not vary significantly across genres.")
```
