````python
¡Por supuesto\! Para que veas cómo integrar el código y las explicaciones en tu proyecto, te muestro la estructura que debe seguir tu **Jupyter/Google Colab Notebook** (`sprint_6_Vacio_notebook.ipynb`).

El formato ideal combina celdas de **Markdown** para los encabezados y las conclusiones (tal como lo has visto en los sprints anteriores) y celdas de **Código** para la ejecución de Pandas, Matplotlib y SciPy.

A continuación, se detalla el contenido de tu notebook, paso a paso:

-----

# Proyecto Integrado 1: Análisis de Éxito de Videojuegos (2017 Planning)

Trabajas para la tienda online **Ice** en diciembre de 2016 y debes identificar patrones de éxito para planificar la campaña publicitaria de 2017.

## Paso 1. Apertura del Archivo de Datos y Estudio de la Información General

En esta sección se cargan las librerías necesarias y se realiza la inspección inicial de los datos.

### 1.1. Inicialización y Carga de Datos

```python
# Importar librerías
import pandas as pd
import numpy as np
from scipy import stats # Para las pruebas de hipótesis (Paso 5)
import matplotlib.pyplot as plt
import seaborn as sns # Para visualizaciones avanzadas (diagramas de caja, dispersión)

# Configuración de visualización
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Cargar el archivo de datos
url = 'https://practicum-content.s3.us-west-1.amazonaws.com/datasets/games.csv'
data = pd.read_csv(url)

# Mostrar las primeras 5 filas e información general
print("Primeras filas del dataset:")
display(data.head()) # Usamos display() para mejor visualización en notebooks

print("\nInformación general del dataset:")
data.info()
```

-----

## Paso 2. Preparación de los Datos

Esta sección se enfoca en la limpieza, la corrección de tipos de datos y el cálculo de ventas totales.

### 2.1. Reemplazo de Nombres de Columnas y Tipos de Datos

```python
# 1. Reemplazar nombres de columnas (minúsculas)
data.columns = data.columns.str.lower()
print("Nuevas columnas:", data.columns.tolist())

# 2. Conversión de tipos de datos y manejo de 'tbd'
# Convertir 'tbd' en 'user_score' a NaN para permitir la conversión a float
data['user_score'] = data['user_score'].replace('tbd', np.nan)
data['user_score'] = data['user_score'].astype(float)

# Explicación de cambio de tipos:
# Se convirtió 'user_score' a float para realizar operaciones matemáticas.
# Las demás columnas de scores ya son numéricas. 'year_of_release' se convertirá a entero después de manejar NaNs.
```

### 2.2. Manejo de Valores Ausentes (NaN)

```python
# Mostrar el porcentaje de valores ausentes por columna
print("\nPorcentaje de valores ausentes:")
print(data.isnull().sum() / len(data) * 100)

# Estrategias de manejo:

# 1. Eliminar filas con NaN en 'name' y 'genre' (menos del 2%, cruciales)
data.dropna(subset=['name', 'genre'], inplace=True)

# 2. Eliminar filas con NaN en 'year_of_release' (crucial para el análisis temporal)
data.dropna(subset=['year_of_release'], inplace=True)
data['year_of_release'] = data['year_of_release'].astype(int)

# 3. Rellenar 'rating' (ESRB) con 'unknown'
data['rating'] = data['rating'].fillna('unknown')

# 4. Dejar NaN en 'critic_score' y 'user_score' para no sesgar las métricas
# (se tratarán durante la correlación y las pruebas de hipótesis).

# Verificación final
data.info()
```

### 2.3. Cálculo de Ventas Totales

```python
# Sumar ventas regionales en una nueva columna 'global_sales'
sales_columns = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
data['global_sales'] = data[sales_columns].sum(axis=1)

print("\nVerificación de 'global_sales' (primeras 5 filas):")
display(data[['name', 'platform', 'global_sales']].head())
```

-----

## Paso 3. Análisis de Datos

### 3.1. Juegos Lanzados por Año

```python
games_per_year = data.groupby('year_of_release')['name'].count()

# Graficar la distribución
plt.figure(figsize=(12, 5))
games_per_year.plot(kind='bar')
plt.title('Número de Juegos Lanzados por Año')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Cantidad de Juegos')
plt.show()

# Conclusión sobre la significancia de los datos:
# (Escribir una breve explicación sobre el pico de lanzamientos y la incompletitud de los datos de 2016 y años muy antiguos).
```

### 3.2. Elección del Período Relevante y Plataformas Líderes

```python
# Determinar período relevante: 2013 en adelante (última generación de consolas)
data_relevant = data[data['year_of_release'] >= 2013].copy()
print(f"Número de juegos en el período relevante (2013+): {data_relevant.shape[0]}")

# Identificar plataformas líderes en el período 2013+
top_platforms_relevant = data_relevant.groupby('platform')['global_sales'].sum().sort_values(ascending=False).head(5).index.tolist()
print("\nTop 5 Plataformas Líderes (2013+):", top_platforms_relevant)

# Distribución de ventas anuales para las plataformas líderes
sales_by_platform_year = data_relevant.groupby(['year_of_release', 'platform'])['global_sales'].sum().reset_index()

plt.figure(figsize=(15, 8))
for platform in top_platforms_relevant:
    platform_data = sales_by_platform_year[sales_by_platform_year['platform'] == platform]
    plt.plot(platform_data['year_of_release'], platform_data['global_sales'], label=platform, marker='o')

plt.title('Ventas Globales Anuales de Plataformas Líderes (2013-2016)')
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Ventas Globales (en millones de USD)')
plt.legend()
plt.grid(True)
plt.show()

# Conclusiones:
# (Escribir una breve explicación sobre qué plataformas crecen (PS4, XOne) y cuáles se reducen (PS3, X360)).
```

### 3.3. Diagrama de Caja: Ventas por Plataforma

```python
# Filtrar solo las plataformas relevantes para el boxplot
relevant_platforms_data = data_relevant[data_relevant['platform'].isin(top_platforms_relevant)]

plt.figure(figsize=(12, 6))
# Boxplot de ventas globales por plataforma (limitando el eje y para ver la distribución central)
sns.boxplot(x='platform', y='global_sales', data=relevant_platforms_data)
plt.ylim(0, 2) # Limitar a 2 millones de ventas para ignorar atípicos extremos
plt.title('Distribución de Ventas Globales por Plataforma (Juegos Individuales, 2013+)')
plt.xlabel('Plataforma')
plt.ylabel('Ventas Globales (en millones de USD)')
plt.show()

# Conclusiones:
# (Explicar las diferencias en ventas promedio (la línea dentro de la caja) y la concentración de los valores atípicos (éxitos)).
```

### 3.4. Correlación entre Reseñas y Ventas (Plataforma Líder: PS4)

```python
# Analizar PS4 como plataforma líder
ps4_data = data_relevant[data_relevant['platform'] == 'PS4'].dropna(subset=['user_score', 'critic_score'])

# Gráficos de dispersión
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x='critic_score', y='global_sales', data=ps4_data)
plt.title('PS4: Reseñas de la Crítica vs Ventas')

plt.subplot(1, 2, 2)
sns.scatterplot(x='user_score', y='global_sales', data=ps4_data)
plt.title('PS4: Reseñas de Usuarios vs Ventas')
plt.show()

# Cálculo de la Correlación
critic_corr = ps4_data['critic_score'].corr(ps4_data['global_sales'])
user_corr = ps4_data['user_score'].corr(ps4_data['global_sales'])

print(f"Correlación PS4: Crítica vs Ventas: {critic_corr:.2f}")
print(f"Correlación PS4: Usuario vs Ventas: {user_corr:.2f}")

# Conclusiones:
# (Comparar el impacto de la crítica frente a la del usuario en las ventas).
```

### 3.5. Distribución de Ventas por Género

```python
# Agrupar y calcular ventas totales y promedio por género (2013+)
genre_analysis = data_relevant.groupby('genre').agg(
    total_sales=('global_sales', 'sum'),
    average_sales=('global_sales', 'mean'),
    game_count=('name', 'count')
).sort_values(by='total_sales', ascending=False)

print("\nAnálisis de Géneros (Ventas Totales y Promedio, 2013+):")
display(genre_analysis)

# Gráfico de barras de Ventas Totales por Género
plt.figure(figsize=(12, 6))
genre_analysis['total_sales'].plot(kind='bar')
plt.title('Ventas Globales Totales por Género (2013+)')
plt.ylabel('Ventas (en millones de USD)')
plt.show()

# Conclusiones:
# (Identificar los géneros más rentables por ventas totales y por promedio por juego).
```

-----

## Paso 4. Creación de un Perfil de Usuario para Cada Región

### 4.1. Top 5 Plataformas y Géneros

```python
regions = ['na_sales', 'eu_sales', 'jp_sales']

for region in regions:
    print(f"\n--- Análisis Regional: {region.upper().replace('_SALES', '')} ---")

    # Top 5 Plataformas
    top_5_platform = data_relevant.groupby('platform')[region].sum().sort_values(ascending=False).head(5)
    print(f"Top 5 Plataformas:\n{top_5_platform}")

    # Top 5 Géneros
    top_5_genre = data_relevant.groupby('genre')[region].sum().sort_values(ascending=False).head(5)
    print(f"Top 5 Géneros:\n{top_5_genre}")
```

### 4.2. Efecto de la Clasificación ESRB en las Ventas

```python
# Agrupar por 'rating' y sumar ventas regionales
esrb_analysis = data_relevant.groupby('rating')[regions].sum()

print("\n--- Efecto de la Clasificación ESRB ---")
display(esrb_analysis)

# Graficar
esrb_analysis[regions].plot(kind='bar', figsize=(10, 6))
plt.title('Ventas por Clasificación ESRB y Región (2013+)')
plt.ylabel('Ventas (en millones de USD)')
plt.show()

# Conclusiones:
# (Comparar la preferencia de Ratings M/E/T en NA/EU frente a JP).
```

-----

## Paso 5. Prueba de Hipótesis

### 5.1. Hipótesis 1: Xbox One vs PC (Calificaciones de Usuario)

```python
# Filtrar datos de las plataformas (excluyendo NaNs)
xbox_scores = data_relevant[(data_relevant['platform'] == 'XOne') & (data_relevant['user_score'].notnull())]['user_score']
pc_scores = data_relevant[(data_relevant['platform'] == 'PC') & (data_relevant['user_score'].notnull())]['user_score']

# Establecer alfa (umbral)
alpha = 0.05

# Formulación de Hipótesis
print("H0: Las calificaciones promedio de usuario para Xbox One y PC son iguales.")
print("H1: Las calificaciones promedio de usuario para Xbox One y PC son diferentes.")

# Prueba T de Student (equal_var=False asume varianzas desiguales)
t_stat, p_value = stats.ttest_ind(xbox_scores, pc_scores, equal_var=False)

print(f"\nValor P: {p_value:.4f}")

# Criterio y Conclusión
if p_value < alpha:
    print(f"Resultado: El valor P ({p_value:.4f}) es menor que alfa ({alpha}). Se RECHAZA la hipótesis nula.")
else:
    print(f"Resultado: El valor P ({p_value:.4f}) es mayor que alfa ({alpha}). NO se puede rechazar la hipótesis nula.")
```

### 5.2. Hipótesis 2: Géneros de Acción vs Deportes (Calificaciones de Usuario)

```python
# Filtrar datos de los géneros (excluyendo NaNs)
action_scores = data_relevant[(data_relevant['genre'] == 'Action') & (data_relevant['user_score'].notnull())]['user_score']
sports_scores = data_relevant[(data_relevant['genre'] == 'Sports') & (data_relevant['user_score'].notnull())]['user_score']

# Establecer alfa (umbral)
alpha = 0.05

# Formulación de Hipótesis
print("H0: Las calificaciones promedio de usuario para los géneros Acción y Deportes son iguales.")
print("H1: Las calificaciones promedio de usuario para los géneros Acción y Deportes son diferentes.")

# Prueba T de Student (equal_var=False asume varianzas desiguales)
t_stat, p_value = stats.ttest_ind(action_scores, sports_scores, equal_var=False)

print(f"\nValor P: {p_value:.4f}")

# Criterio y Conclusión
if p_value < alpha:
    print(f"Resultado: El valor P ({p_value:.4f}) es menor que alfa ({alpha}). Se RECHAZA la hipótesis nula.")
else:
    print(f"Resultado: El valor P ({p_value:.4f}) es mayor que alfa ({alpha}). NO se puede rechazar la hipótesis nula.")
```

-----

## Paso 6. Conclusión General

**(Esta es una celda de Markdown)**

Aquí debes resumir tus hallazgos clave y proporcionar una recomendación estratégica a la tienda Ice para su campaña publicitaria de 2017.

**Resumen de Hallazgos Clave:**

  * **Periodo de Relevancia:** (Mencionar el período 2013-2016 y por qué se eligió).
  * **Plataformas:** (Identificar PS4 y XOne como las más prometedoras y por qué).
  * **Géneros:** (Identificar Action/Shooter como los más rentables).
  * **Influencia de Reseñas:** (Resaltar que la Crítica tiene mayor correlación con las ventas que la opinión del Usuario).
  * **Perfiles Regionales:** (Señalar las claras diferencias entre NA/EU y JP).
  * **Pruebas de Hipótesis:** (Indicar el resultado de cada prueba: si se acepta o se rechaza H0).

**Recomendación Estratégica para 2017:**

  * **Enfoque de Plataforma y Género:** La inversión debe centrarse en juegos de **Shooter** y **Action** para **PS4** y **XOne**.
  * **Marketing:** El éxito comercial debe predecirse basándose en las **reseñas de expertos** (Crítica) en lugar de las calificaciones promedio de los usuarios.
  * **Ajuste Regional:** Las campañas deben ser **diferentes** en Japón, enfocándose en plataformas como **3DS** y géneros como **Role-Playing**.
````
