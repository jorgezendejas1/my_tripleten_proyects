```python
import pandas as pd

# Load the dataset
df = pd.read_csv('https://practicum-content.s3.us-west-1.amazonaws.com/datasets/games.csv')

# Display the first few rows
print("First 5 rows of the DataFrame:")
display(df.head())

# Display general information about the DataFrame
print("\nGeneral information about the DataFrame:")
display(df.info())
```

```python
# Reemplazar nombres de columnas a minúsculas
df.columns = df.columns.str.lower()

# Convertir 'user_score' a numérico, manejando 'TBD' y valores ausentes
# 'TBD' se puede considerar como un valor ausente ya que no es una calificación numérica.
# Convertiremos los valores a numéricos y los errores se convertirán en NaN.
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')

# Convertir 'year_of_release' a entero. Los valores ausentes (NaN) se mantendrán como NaN.
# Explicación: Convertimos a entero para que represente el año de lanzamiento de forma discreta.
# Los valores ausentes podrían deberse a que la información no estaba disponible en el momento de la recopilación.
# Decidimos dejarlos como NaN por ahora, ya que rellenarlos podría introducir sesgos si no tenemos una base sólida para la imputación.
df['year_of_release'] = df['year_of_release'].astype('Int64') # Usamos Int64 para permitir NaNs en una columna entera

# Manejar valores ausentes en otras columnas si es necesario.
# Para 'name' y 'genre', los valores ausentes podrían indicar errores en la entrada de datos.
# Podríamos eliminar estas filas si son pocas, o rellenarlas con un marcador como 'Unknown'.
# Por ahora, los dejaremos como están para no perder otros datos.
# Para 'critic_score' y 'rating', los valores ausentes son comunes en las reseñas y clasificaciones.
# Dejarlos como NaN es apropiado ya que no tenemos información para rellenarlos de manera significativa.

# Explicación de los cambios de tipo de datos:
# - 'user_score': Cambiado de 'object' a numérico (float64 después de coerce) para permitir cálculos numéricos.
# - 'year_of_release': Cambiado de 'float64' a 'Int64' para representar años como enteros y manejar NaNs.

# Calcular las ventas totales
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']

# Mostrar información actualizada del DataFrame
print("\nInformación del DataFrame después de la preparación de datos:")
display(df.info())

# Mostrar las primeras filas con la nueva columna 'total_sales'
print("\nPrimeras 5 filas con ventas totales:")
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Analizar cuántos juegos fueron lanzados en diferentes años
games_by_year = df.groupby('year_of_release')['name'].count()

print("Número de juegos lanzados por año:")
display(games_by_year)

# Graficar la distribución de juegos por año
plt.figure(figsize=(12, 6))
games_by_year.plot(kind='bar')
plt.title('Número de juegos lanzados por año')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Número de juegos')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Analizar cómo varían las ventas de una plataforma a otra a lo largo del tiempo
# Agrupar por año y plataforma y sumar las ventas totales
platform_sales_by_year = df.groupby(['year_of_release', 'platform'])['total_sales'].sum().reset_index()

print("\nVentas totales por plataforma y año:")
display(platform_sales_by_year.head())

# Identificar las plataformas con mayores ventas totales históricas para un análisis inicial
top_platforms_total_sales = df.groupby('platform')['total_sales'].sum().sort_values(ascending=False)

print("\nPlataformas con mayores ventas totales históricas:")
display(top_platforms_total_sales.head(10))

# Para un análisis más detallado y relevante para 2017, nos enfocaremos en datos más recientes.
# Basándonos en la información general (df.info()), los datos llegan hasta 2016.
# Consideraremos un período relevante para el análisis, por ejemplo, los últimos 10 años con datos completos (hasta 2016).
# Esto nos ayuda a entender las tendencias recientes y la vida útil de las plataformas.

# Determinar el período relevante para el análisis (ej. últimos 10 años con datos hasta 2016)
# En este caso, consideraremos el período de 2007 a 2016 como un ejemplo de un período significativo con datos.
# Se puede ajustar este rango según la observación del gráfico de juegos por año y el ciclo de vida de las plataformas.
recent_years_df = df[(df['year_of_release'] >= 2007) & (df['year_of_release'] <= 2016)].copy()

# Analizar las ventas de las plataformas principales en el período reciente
top_platforms_recent = recent_years_df.groupby('platform')['total_sales'].sum().sort_values(ascending=False).head(10).index.tolist()

print(f"\nLas 10 plataformas principales en ventas totales en el período 2007-2016 son: {top_platforms_recent}")

# Construir un gráfico de la distribución global de los juegos por años para estas plataformas principales
plt.figure(figsize=(14, 8))
for platform in top_platforms_recent:
    platform_data = recent_years_df[recent_years_df['platform'] == platform]
    platform_sales_by_year_recent = platform_data.groupby('year_of_release')['total_sales'].sum()
    plt.plot(platform_sales_by_year_recent.index, platform_sales_by_year_recent.values, label=platform)

plt.title('Ventas totales globales por año para las plataformas principales (2007-2016)')
plt.xlabel('Año de lanzamiento')
plt.ylabel('Ventas totales globales (millones)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Analizar las 5 plataformas principales por región
regions = ['na_sales', 'eu_sales', 'jp_sales']
region_names = ['Norteamérica', 'Europa', 'Japón']

for region, region_name in zip(regions, region_names):
    print(f"\nAnalizando las 5 plataformas principales en {region_name}:")
    top_platforms_region = df.groupby('platform')[region].sum().sort_values(ascending=False).head(5)
    display(top_platforms_region)

# Analizar los 5 géneros principales por región
for region, region_name in zip(regions, region_names):
    print(f"\nAnalizando los 5 géneros principales en {region_name}:")
    top_genres_region = df.groupby('genre')[region].sum().sort_values(ascending=False).head(5)
    display(top_genres_region)
```

```python
# Analizar si las clasificaciones ESRB afectan a las ventas en regiones individuales
regions = ['na_sales', 'eu_sales', 'jp_sales']
region_names = ['Norteamérica', 'Europa', 'Japón']

for region, region_name in zip(regions, region_names):
    print(f"\nAnalizando el impacto de las clasificaciones ESRB en las ventas en {region_name}:")
    # Agrupar por clasificación ESRB y sumar las ventas de la región
    sales_by_rating = df.groupby('rating')[region].sum().sort_values(ascending=False)
    display(sales_by_rating)

    # Opcional: Graficar la distribución de ventas por clasificación ESRB
    plt.figure(figsize=(8, 5))
    sales_by_rating.plot(kind='bar')
    plt.title(f'Ventas totales por clasificación ESRB en {region_name}')
    plt.xlabel('Clasificación ESRB')
    plt.ylabel(f'Ventas totales ({region})')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```

```python
from scipy import stats as st

# Hipótesis 1: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.
# Hipótesis nula (H₀): Las calificaciones promedio de los usuarios para Xbox One y PC son iguales (μ₁ = μ₂).
# Hipótesis alternativa (H₁): Las calificaciones promedio de los usuarios para Xbox One y PC son diferentes (μ₁ ≠ μ₂).

# Filtra los datos para las plataformas Xbox One y PC, excluyendo los valores NaN en user_score
xone_scores = df[(df['platform'] == 'XOne') & (df['user_score'].notna())]['user_score']
pc_scores = df[(df['platform'] == 'PC') & (df['user_score'].notna())]['user_score']

# Establecer el valor de umbral alfa
alpha = 0.05

# Realizar la prueba t de Student para dos muestras independientes
# equal_var=False porque no podemos asumir que las varianzas de las poblaciones son iguales.
ttest_xone_pc = st.ttest_ind(xone_scores, pc_scores, equal_var=False)

print("Prueba de hipótesis: Calificaciones promedio de usuarios para Xbox One y PC")
print(f"Estadístico t: {ttest_xone_pc.statistic:.4f}")
print(f"Valor p: {ttest_xone_pc.pvalue:.4f}")

# Evaluar el resultado
if ttest_xone_pc.pvalue < alpha:
    print(f"El valor p ({ttest_xone_pc.pvalue:.4f}) es menor que el umbral alfa ({alpha}).")
    print("Rechazamos la hipótesis nula: Las calificaciones promedio de los usuarios para Xbox One y PC son diferentes.")
else:
    print(f"El valor p ({ttest_xone_pc.pvalue:.4f}) es mayor o igual que el umbral alfa ({alpha}).")
    print("No rechazamos la hipótesis nula: Las calificaciones promedio de los usuarios para Xbox One y PC son las mismas.")

# Hipótesis 2: Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.
# Hipótesis nula (H₀): Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son iguales (μ₁ = μ₂).
# Hipótesis alternativa (H₁): Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son diferentes (μ₁ ≠ μ₂).

# Filtra los datos para los géneros Acción y Deportes, excluyendo los valores NaN en user_score
action_scores = df[(df['genre'] == 'Action') & (df['user_score'].notna())]['user_score']
sports_scores = df[(df['genre'] == 'Sports') & (df['user_score'].notna())]['user_score']

# Realizar la prueba t de Student para dos muestras independientes
# equal_var=False porque no podemos asumir que las varianzas de las poblaciones son iguales.
ttest_action_sports = st.ttest_ind(action_scores, sports_scores, equal_var=False)

print("\nPrueba de hipótesis: Calificaciones promedio de usuarios para los géneros Acción y Deportes")
print(f"Estadístico t: {ttest_action_sports.statistic:.4f}")
print(f"Valor p: {ttest_action_sports.pvalue:.4f}")

# Evaluar el resultado
if ttest_action_sports.pvalue < alpha:
    print(f"El valor p ({ttest_action_sports.pvalue:.4f}) es menor que el umbral alfa ({alpha}).")
    print("Rechazamos la hipótesis nula: Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son diferentes.")
else:
    print(f"El valor p ({ttest_action_sports.pvalue:.4f}) es mayor o igual que el umbral alfa ({alpha}).")
    print("No rechazamos la hipótesis nula: Las calificaciones promedio de los usuarios para los géneros Acción y Deportes son las mismas.")
```
